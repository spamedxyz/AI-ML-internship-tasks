# Task 1: Exploring and Visualizing the Iris Dataset

**Objective:** Load, inspect, and visualize the Iris dataset to understand data trends and distributions.

**Skills:** Data loading with pandas, descriptive statistics, matplotlib/seaborn plotting.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Load the Dataset

```python
# Load Iris dataset via seaborn (built-in CSV equivalent)
df = sns.load_dataset('iris')
print('Shape:', df.shape)
print('\nColumn names:', df.columns.tolist())
df.head()
```

## 2. Data Inspection

```python
df.info()
```

```python
df.describe()
```

```python
print('Missing values:\n', df.isnull().sum())
print('\nSpecies counts:\n', df['species'].value_counts())
```

## 3. Visualizations

### 3.1 Scatter Plots — Feature Relationships

```python
sns.pairplot(df, hue='species', palette='Set2', diag_kind='hist')
plt.suptitle('Iris Feature Relationships by Species', y=1.02)
plt.show()
```

```python
fig, ax = plt.subplots()
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', palette='Set2', ax=ax)
ax.set_title('Sepal Length vs Petal Length')
plt.show()
```

### 3.2 Histograms — Value Distributions

```python
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df[numeric_cols].hist(bins=15, figsize=(12, 8), edgecolor='black')
plt.suptitle('Distribution of Iris Features', y=1.02)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()
for i, col in enumerate(numeric_cols):
    sns.histplot(data=df, x=col, hue='species', kde=True, ax=axes[i], palette='Set2')
    axes[i].set_title(f'{col} by Species')
plt.tight_layout()
plt.show()
```

### 3.3 Box Plots — Outlier Detection

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(data=df, x='species', y='sepal_width', palette='Set2', ax=ax)
ax.set_title('Sepal Width by Species (Outliers)')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()
for i, col in enumerate(numeric_cols):
    sns.boxplot(data=df, x='species', y=col, palette='Set2', ax=axes[i])
    axes[i].set_title(col)
plt.suptitle('Box Plots by Species', y=1.02)
plt.tight_layout()
plt.show()
```

## 4. Results and Insights

- The dataset has **150 rows** and **4 numeric features** plus a **species** label (3 classes).
- **Setosa** is clearly separated on petal length/width; **versicolor** and **virginica** overlap somewhat.
- **Sepal width** shows a few mild outliers per species in box plots.
- Petal measurements are the strongest discriminators between species.
